### Setting up random seed 

```python
np.random.seed(1)
```

### Seperating the data and Labels

```python
X = df.drop('feature_60' , axis = 1)
y = df['feature_60']
print(X)
print(y)
```

### Splitting to training and test data 

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.1 , stratify= y)
```

```python
# check shape 
print(X.shape , X_train.shape , X_test.shape)
```

### Instantiate and fit the model 

```python
model = LogisticRegression()
model.fit(X_train , y_train)
```

### Model evaluation

Accuracy score of training data : 
1. Firstly we take prediction of what our X_train data will give
2. we use .predict() for doint this
3. then we check the accuracy score by comparing it with y_train
4. that is our label and we use the accuracy_score() function for this 

```python
# accuracy on training model 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , y_train)

print(f"aAccuracy score on the traininng data is : {training_data_accuracy}")
```

Accuracy score on test data : 

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , y_test)

print(f"Accuracy score on the test data is : {test_data_accuracy}")
```
